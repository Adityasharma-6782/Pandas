```python
import numpy as np
import pandas as pd
```

Missing Data

```python
# Finding missing data
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],             # np.nan ==> NULL Value
    'B': [np.nan, 2, 3, 4, 5],            
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()    # NULL values pe True
```

```python
df.isna().sum()  # It count total NULL values in a column
```

```python
df.isna().any()     # column me at least ek null value hai to true
```

```python
#  Removing Missing Data
#     based on rows
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],             # np.nan ==> NULL Value
    'B': [1, 2, 3, 4, 5],            
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df1 = pd.DataFrame(data)
```

```python
df1
```

```python
df1.dropna()  # return karta hai jisme koi bhi NULL values nhi ho
```

```python
df1.dropna(thresh=3)   # at least 3 not null values in a row
```

```python
df1.dropna(thresh=4)
```

```python
# Filling the missing data
```

```python
df1.fillna(100)
```

```python
df1  # ==> parmanent ke liye inplace=True likhna padega
```
