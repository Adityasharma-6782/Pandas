**Aggregation and GroupBy**

GroupBy

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by categories and calculate the sum od series
cat = df.groupby('Category')
cat   # It is a object
for i, v in cat:
    print(i)
    print(v)
```

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by stote and calculate the sum of Sales
sto = df.groupby('Store')['Sales']
sto
for i,v in sto:
    print(i)
    print(v)
```

```python
sto = df.groupby('Store')['Sales'].sum()
sto
```

```python
# Group multiple columns
mul = df.groupby(['Category','Store'])
mul
for i, v in mul:
    print(i)
    print(v)
```

```python
mul = df.groupby(['Category','Store'])['Sales'].sum()
mul
```
